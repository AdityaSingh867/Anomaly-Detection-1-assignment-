Q3. How does unsupervised anomaly detection differ from supervised anomaly detection?

Actually in supervised learning, you have the data set labelled and you pass the labelled values as you train the model so that it learns parameters that will separate the 'good' from 'bad' results. In anomaly detection, it is unsupervised as you do not pass any labelled values.
